# R√©gression lin√©aire

```python
X= avocado.iloc[0:52,2]
Y= avocado.iloc[0:52,7]
```

```python
axes =plt.axes()
axes.grid()
plt.scatter(X,Y)
plt.title("Corr√©lation lin√©aire entre le prix moyen et le total de sac(en 2015,en Albany)")
plt.show()
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
```

```python
def predict(X):
   return slope * X + intercept
```

```python
fitLine = predict(X)
plt.plot(X, fitLine, c='r')
```
